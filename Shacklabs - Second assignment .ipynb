Record linking and fuzzy matching are terms used to describe the process of joining two data sets together that do not 
have a common unique identifier.This is a common business challenge wherein common key to merge dataset might not be available . 

Convention tools such as excel sheet by using vlookup etc. might not be clean and also ineffective if the volume is too high.Also the common identifier or a key is missing in this case . There are couple of option available such as fuzzywuzzy alogrithm however those are computationaly intensive . 

We have decided to approach this with python library called fuzzymatcher as the libray works well with hugh volume and interface
is clean and easy to understand. 

```python
import pandas as pd 
import fuzzymatcher
```

```python
df_amazon = pd.read_csv("amz_com-ecommerce_sample.csv",encoding='latin-1')
```

```python
df_amazon.head()
```

```python
df_flipkart = pd.read_csv("flipkart_com-ecommerce_sample.csv")
```

```python
df_flipkart.head()
```

```python
# Here is where you will put the key column name on which you would like your datasets to get matched.
left_on = ["product_name"]
right_on = ["product_name"]
```

```python
# Below is how you join the dataset . REemeber that we are not dropping columns but the match would happen and a score would be 
# assigned to the match between two key parameters ,bigger the score , better is the match  . AFter the match is complete 
#we can pull # whatever column we require to furnish the data set.
results = fuzzymatcher.fuzzy_left_join(df_amazon,df_flipkart,left_on,right_on)
```

```python
results.head(5)
```

```python
results.columns
```

```python
# REname the columns to understand the result dataset properly.
results.rename(columns={'product_name_left':'Product name in Amazon','retail_price_left':'Retail price in Amazon','product_name_right':'Product name in Flipkart',
                        'retail_price_right':'Retail_price in Flipkart','discounted_price_left':'Discounted price in Amazon',
                        'discounted_price_right':'Discounted price in Flipkart'},inplace=True)
```

```python
cols = ['best_match_score','Product name in Flipkart','Retail_price in Flipkart','Discounted price in Flipkart','Product name in Amazon',
        'Retail price in Amazon','Discounted price in Amazon']
```

As mentioned before Best match score will be assigned to each and evey match between the two key of the dataset . This scores 
can than be used within the result dataset to showcase the information from the dataset ,remember higher the score , better is the match.

```python
results[cols].sort_values(by=['best_match_score'], ascending=False).head(10)
```
